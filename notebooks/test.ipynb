```python
from d2s import plot, dice, Bonus
```

```python
# Axt vs Baron
plot.fight(
    attack_dice=(dice.blue(), dice.red()),
    defense_dice=(dice.brown(), dice.gray(), dice.black()),
    bonus=Bonus(damage=1, surge=1, pierce=1),
    surge_bonuses=(Bonus(damage=2), Bonus(pierce=1)),
)
```

```python
# Bogen vs Baron
plot.fight(
    attack_dice=(dice.blue(), dice.red(), dice.yellow()),
    defense_dice=(dice.brown(), dice.gray(), dice.black()),
    bonus=Bonus(range=4),
    surge_bonuses=(Bonus(damage=2),),
)
```

```python
# Zombie vs Baron
plot.fight(
    attack_dice=(dice.blue(), dice.red(), dice.yellow()),
    defense_dice=(dice.brown(), dice.gray(), dice.black()),
    bonus=Bonus(),
    surge_bonuses=(Bonus(damage=1),),
)
```

```python
# Axt vs Drache
plot.fight(
    attack_dice=(dice.blue(), dice.red()),
    defense_dice=(dice.gray(), dice.gray()),
    bonus=Bonus(damage=1, surge=1, pierce=1),
    surge_bonuses=(Bonus(damage=2), Bonus(pierce=1)),
)
```
